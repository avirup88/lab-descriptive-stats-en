#### Show a visual summary of the data using boxplot. What Are the First and Third Quartiles? Comment your results.

* Hint : Quantiles vs Quartiles*
In Probability and Statistics, quantiles are cut points dividing the range of a probability distribution into continuous intervals with equal probabilities. When division is into four parts the values of the variate corresponding to 25%, 50% and 75% of the total distribution are called quartiles.

```python
# YOUR CODE HERE
# Creating boxplots for numerical columns
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].boxplot(employee['Years'], patch_artist=True, boxprops=dict(facecolor='lightblue', color='black'))
axs[0].set_title('Boxplot of Years')
axs[0].set_ylabel('Years')

axs[1].boxplot(employee['Salary'], patch_artist=True, boxprops=dict(facecolor='lightgreen', color='black'))
axs[1].set_title('Boxplot of Salary')
axs[1].set_ylabel('Salary')

plt.tight_layout()
plt.show()

# Calculating the first and third quartiles for each column
years_q1, years_q3 = np.percentile(employee['Years'], [25, 75])
salary_q1, salary_q3 = np.percentile(employee['Salary'], [25, 75])
```

```python
# print first quartile here
print("First and Third Quartiles:")
print(f"Years Q1: {years_q1}")
print(f"Salary Q1: {salary_q1}")
```

```python
# print third quartile here
print("First and Third Quartiles:")
print(f"Years Q3: {years_q3}")
print(f"Salary Q3: {salary_q3}")
```
